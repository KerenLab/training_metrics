```python
import os
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
markers_df = pd.read_csv(r'C:\Users\leeatgen.WISMAIN\Desktop\Vova\Leeat Keren\wet_experiments\markers and datasets\markers_array.csv')
```

```python
markers_names = markers_df.loc[:,'Name']
```

```python
markers_dict = Counter(markers_names)
```

```python
sorted_markers = dict(sorted(markers_dict.items(), key=lambda item: item[1], reverse = True))
```

```python
top50 = dict()
i = 0
for key,value in sorted_markers.items():
    if i == 60:
        break
    top50[key] = value
    i+=1
```

```python
import seaborn as sns
sns.set()
plt.figure(figsize = (50,10))
plt.bar(top50.keys(), top50.values())
plt.yticks(range(1,13));
```

```python
top50csv = pd.DataFrame.from_dict(top50.keys())
```

```python
top50csv.to_csv(r'C:\Users\leeatgen.WISMAIN\Desktop\Vova\Leeat Keren\wet_experiments\markers and datasets\top50.csv')
```
